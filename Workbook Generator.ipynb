```python
import glob
from openpyxl import Workbook
from openpyxl import load_workbook
import pandas as pd
```

```python
files_predictor = glob.glob('/content/datasets/MyDrive/Datasets/GCM Data (Daily)/CSV Files' + '/*.csv')
```

```python
files_rainfall = glob.glob('/content/datasets/MyDrive/Datasets/IMD Rainfall 0.25x0.25/NC Files/csv files' + '/*.csv')
```

```python
files_rainfall = sorted(files_rainfall)
```

```python
files_rainfall
```

```python
for count, (f1,f2) in enumerate(zip(files_predictor, files_rainfall)):
  wb = Workbook()
  (f_path, f_name) = os.path.split(f1)
  (f_short_name, f_extension) = os.path.splitext(f_name)
  wb.save(filename = f_short_name)
  temp_sheet_pred = pd.read_csv(f1)
  temp_sheet_pred
  temp_sheet_rain = pd.read_csv(f2)
  writer = pd.ExcelWriter(f_short_name + '.xlsx')
  temp_sheet_pred.to_excel(writer, sheet_name= f_short_name + ' predictors')
  temp_sheet_rain.to_excel(writer, sheet_name= f_short_name + ' rainfall')
  writer.save()
  print(count)

  
  #wb = Workbook()
  #(f_path, f_name) = os.path.split(filename)
  #(f_short_name, f_extension) = os.path.splitext(f_name)
  #ws = wb.add_sheet(f_short_name)




```
